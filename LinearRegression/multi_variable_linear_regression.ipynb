```python
import tensorflow as tf
import numpy as np
X = np.array([[73., 80., 75.],[93., 88., 93.],[89., 91., 90.],
          [96., 98., 100.],[73., 66., 70.]], dtype = np.float32)
Y = np.array([[152.],[185.],[180.],[196.],[142.]], dtype = np.float32)

#X = x_data[:, :]
#Y = y_data[:, :]

# W는 3*1 행렬
W = tf.Variable(tf.random.normal([3,1]))
# b는 1*1 헹렬
b = tf.Variable(tf.random.normal([1]))

#XW+b 행렬의 곱으로 표현
def multiLinear(x):
    return tf.matmul(X, W) + b

optimizer = tf.optimizers.SGD(learning_rate = 1e-5)

for step in range(2001):
    with tf.GradientTape() as g:
        hypothesis = multiLinear(X)
        cost = tf.reduce_mean(tf.square(hypothesis-Y))
    gradients = g.gradient(cost, [W, b])
    #업데이트
    optimizer.apply_gradients(zip(gradients, [W,b]))
    if step % 10 == 0:
        print("%i cost: %f \nprediction\n"% (step, cost), hypothesis.numpy())  

```
