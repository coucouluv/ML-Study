```python
import tensorflow as tf
import matplotlib.pyplot as plt

X = [1.1,2.2,3.3]
Y = [1.1,2.2,3.3]


# Wx 일차원 함수
def linear(x, feed_W):
    result = []
    for i in x:
        result.append(i*feed_W)
    return result 

# 실제 값과 hypothesis 사이 cost
def findcost(hypothesis, y):
    result = 0
    for i in range(len(y)):
        result += (hypothesis[i]-y[i])**2
    return result/len(y)

W_val = []
cost_val = []

#-3~5까지 W값을 변경해가면서 cost구하기
for i in range(-30, 50):
    feed_W = i * 0.1
    hypothesis = linear(X,feed_W)
    cost = findcost(hypothesis, Y)
    W_val.append(feed_W)
    cost_val.append(cost)
    
#x축은 W값, y축은 cost값으로 함수를 표현 
plt.plot(W_val, cost_val)
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# 방법이 살짝 다른 버전

X = [1,2,3]
Y = [1,2,3]


# Wx 일차원 함수
def linear(x,feed_W):
    return x * feed_W 

# 실제 값과 hypothesis 사이 cost
def findcost(hypothesis, y):
    return tf.reduce_mean(tf.square(hypothesis- y))

W_val = []
cost_val = []
for i in range(-30, 50):
    feed_W = tf.Variable([i * 0.1])
    hypothesis = linear(X, feed_W)
    cost = findcost(hypothesis, Y)
    W_val.append(feed_W)
    cost_val.append(cost)

plt.plot(W_val, cost_val)
plt.show()
```
